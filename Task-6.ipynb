```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\suyog\\Downloads\\disney_plus_titles.csv')
```

```python
df_cleaned = df.fillna('Unknown')
```

```python
# Describe the dataset
description = df_cleaned.describe(include='all')
print(description)
```

```python
# Visualize the distribution of 'release_year'
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['release_year'], kde=True)
plt.title('Distribution of Release Year')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

```python
#graph of the count of different types (Movies vs TV Shows)
plt.figure(figsize=(10, 6))
sns.countplot(data=df_cleaned, x='type')
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
#graph of the correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df_cleaned.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
